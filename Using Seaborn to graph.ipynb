```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


df = pd.read_csv(r'C:\Users\moren\Downloads\Minimum _Wage_Data\Minimum_wage_Data.csv', encoding='Latin')
print(df.head(n=7))
```

```python
## using one Variable 
A = df[:7]
sns.catplot(x="State", y="High.2018", kind="box", data=A);
```

```python

B = df[8:12]
sns.catplot(x="State", y="High.2018", kind="box", data=B)
```

```python
sns.catplot(x="Year", y="High.2018", hue="State", kind="swarm", height=7, data=df);
```

```python

sns.catplot(x="High.Value", y="State", kind="box", height=10, data=df)
```

```python
## multiple Varables
sns.relplot(x="High.Value", y="Year", hue="State", height=8,
            kind="line", estimator=None, data=df);
```

```python
#Low. Value
sns.relplot(x="Low.Value", y="Year", hue="State", height=10,
            kind="line", estimator=None, data=df);
```

```python
sns.relplot(x="Year", y="High.Value", hue="State",
                       col="State", col_wrap= 4,
            height=5, aspect=2, linewidth=2.5,
            kind="line", data=df);
```
